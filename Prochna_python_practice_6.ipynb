# I. Docstrings.

1. Що таке docstrings? Навіщо вони потрібні?

Docstrings - це рядки документації, розміщені в початку модуля, класу, функції чи методу, які надають інформацію про їх призначення та використання. Вони призначені для: документації коду, автоматичної генерації документації,
Підтримки інтерактивних інтерпретаторів, покращення читабельності

4. Створіть docstring для функції, що перевіряє, чи є число простим. Увага: під час виклику функції ми очікуємо, що користувач(ка) може подати на вхід будь-який тип обʼєкту. У випадку, якщо на вході не підходящий тип даних, маємо кинути помилку.

```python
''' Check the number is it prime or not
  Examples
    >>> is_prime(13)
    True
    >>> is_prime(9)
    False
    >>> is_prime(4)
    False

  Args
    num(int): number to check is prime or not

  Returns:
    Boolean: True if number is prime

  Raises
    TypeError: If operand is not an integer
  '''
```

5. Визначте функцію, яка перевіряє, чи є число простим, використовуючи документацію з завдання 4.

```python
def is_prime(num):
  if not (type(num) == int):
    raise TypeError("Print the number to check")

  if num % 2 == 0:
    return False

  for i in range(2, int(num**0.5) + 1):
      if num % i == 0:
          return False

  return True

print("7 is prime: ", is_prime(7))
print("9 is prime: ",is_prime(9))
print("2 is prime: ",is_prime(2))

```

6. Створіть docstring для функції, що повертає найдовший спільний префікс двох рядків не залежно від регістру. Обовʼязково включіть два приклади згідно формату.

```python
''' Returns the longest common prefix of 2 strings
  Examples
    >>> prefix("heelo", "Heelen")
    heel
    >>> prefix("hik", "hok")
    h
    >>> prefix("hello", "world")


  Args
    left_str(str): first string to compare
    right_str(str): second string to compare

  Returns:
    String: the longest common prefix of operands

  Raises
    TypeError: If operand is not a string
  '''
```

7. Реалізуйте функцію згідно документації із завдання 6, використовуючи її.

```python
def prefix(left_str, right_str):
  common_prefix = ""
  min_len = min(len(left_str), len(right_str))

  for i in range(min_len):
      if left_str[i].lower() == right_str[i].lower():
          common_prefix += left_str[i]
      else:
          break

  return common_prefix
print("The longest common prefix for:")
print("heelo and Heelen:", prefix("heelo", "Heelen"))
print("hik and hok:", prefix("hik", "hok"))
print("hello, world:", prefix("hello", "world"))
```

8. Створіть docstring для функції, що розраховує суму щомісячного платежу по єОселі.

```python
''' Calculate the amount of the monthly payment for eOseli
The preferential rate for the entire term of the loan is 3%

  Examples
    >>> calculate_emi(10000, 5, 12)
    846.94
    >>> calculate_emi(10000, 7, 15)
    680.08

  Args
    principal (float): minimum down payment 20%
    number_of_installments (int): number of installments the maximum loan term is 20 years

  Returns:
    Float: the amount of the monthly payment for eOseli

  Raises
    TypeError: If operands are not a number
  '''
```

9. Використовуючи документацію із завдання 8, реалізуйте відповідну функцію.

```python
def calculate_emi(principal, num_of_installments):
  monthly_rate = 3 / (12 * 100)
  emi = (principal * monthly_rate) / (1 - (1 + monthly_rate) ** - num_of_installments)
  return round(emi, 2)

print(calculate_emi(10000, 12))
print(calculate_emi(10000, 15))
```

10. Реалізуйте функцію, яка отримує значення зі словника за ключем та повертає його, а за умови відсутності ключа відловлює KeyError.

```python
def get_by_key(key, input_dict):
  try:
    return input_dict[key]
  except:
    raise KeyError("No such key in a dictionary")

squares = {1: 1, 3: 9, 5: 25, 7: 49, 9: 81}
print(squares)
print("key = 3, value = ", get_by_key(3, squares))

try:
    result = get_by_key(squares, 'b')
    print(result)
except KeyError as e:
    print(f"Error: {e}")
```

11. Реалізуйте функцію із завдання 10 без відловлювання помилки, однак за допомогою конструкції if-else.

```python
def get_by_key(key, input_dict):
  if key in input_dict:
    return input_dict[key]
  else:
    print("No such key in a dictionary")
    return None

squares = {1: 1, 3: 9, 5: 25, 7: 49, 9: 81}
print(squares)
print("key = 3, value = ", get_by_key(3, squares))
print("key = 6, value = ", get_by_key(6, squares))
```

13. Визначте функцію, яка обробляє значення, введене користувачем з терміналу, приводячи його до цілого числа. Використайте відловлення відповідної помилки. Наприклад, людина має ввести свій вік.

```python
def get_age():
  try:
    user_input = input("Type your age: ")
    age = int(user_input)
    return age
  except ValueError:
    print("Error: Type an integer.")
    raise

try:
    user_age = get_age()
    print(f"Your age: {user_age}")
except ValueError:
    print("Error typing an age.")
```

14. Визначте функцію з завдання 13, де використайте конструкцію if-else замість відловдення помилки.

```python
def get_age():
  user_input = input("Type your age: ")

  if user_input.isdigit():
    age = int(user_input)
    return age
  else:
    print("Error: Type an integer.")
    return None


user_age = get_age()
print(f"Your age: {user_age}")
```

15. Визначте функцію з завдання 13, враховуючи додаткове обмеження, що число має бути у діапазоні від 18 до 120 (повнолітня людина). Відловіть відповідні типи помилок.

```python
def get_age():
  user_input = input("Type your age: ")

  if user_input.isdigit():
    age = int(user_input)
    if 18 <= age <= 120:
      return age
    else:
      raise ValueError("Age should be between 18 and 120.")
  else:
    raise ValueError("Age should be an integer.")

# Приклад використання
try:
    user_age = get_age()
    print(f"Your age: {user_age}")
except ValueError:
    print("Error typing an age")
```
